```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
# 데이터 불러오기
# 데이터 파일들이 'data' 폴더 안에 있으므로 경로를 'data/'로 지정해줘야 해.
train_df = pd.read_csv('data/train.csv')
test_df = pd.read_csv('data/test.csv')

# 데이터 기본 정보 확인
print("Train Data Info:")
train_df.info()

print("\nTest Data Info:")
test_df.info()

# 데이터 상위 5개 행 확인
print("\nTrain Data Head:")
print(train_df.head())
```

```python
# 기술 통계량 확인
print("Train Data Describe:")
print(train_df.describe())

print("\nTest Data Describe:")
print(test_df.describe())
```

```python
# 결측치 개수 확인
print("Train Data Missing Values:")
print(train_df.isnull().sum())

print("\nTest Data Missing Values:")
print(test_df.isnull().sum())
```

```python
# stress_score의 분포 확인 (히스토그램)
plt.figure(figsize=(10, 6))
sns.histplot(train_df['stress_score'], kde=True)
plt.title('Distribution of Stress Score')
plt.show()

# 상관관계 히트맵
plt.figure(figsize=(12, 10))
sns.heatmap(train_df.corr(numeric_only=True), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap of Train Data')
plt.show()
```

```python
# 결측치 매트릭스 시각화
msno.matrix(train_df, figsize=(10, 5))
plt.title('Missing Value Matrix of Train Data', fontsize=15)
plt.show()

# 컬럼별 결측치 막대 그래프 시각화
msno.bar(train_df, figsize=(10, 5))
plt.title('Missing Value Bar Chart of Train Data', fontsize=15)
plt.show()

# 결측치 상관관계 히트맵
# 두 컬럼의 결측치 발생 패턴이 얼마나 유사한지 보여줘.
msno.heatmap(train_df, figsize=(10, 5))
plt.title('Missing Value Correlation Heatmap', fontsize=15)
plt.show()
```

```python
# 결측치가 있는 4개 컬럼 리스트
missing_cols = ['medical_history', 'family_medical_history', 'edu_level', 'mean_working']
# 모든 컬럼 리스트 (ID와 stress_score는 제외하고)
all_cols = train_df.columns.drop(['ID', 'stress_score'])

# 각 결측치 컬럼에 대해 모든 컬럼과 비교 분석
for missing_col in missing_cols:
    print(f"\n--- {missing_col} 결측치 분석 ---")
    
    # 결측치 여부를 나타내는 컬럼 생성
    train_df[f'{missing_col}_missing'] = train_df[missing_col].isnull().astype(int)
    
    # 결측치 컬럼을 제외한 나머지 컬럼들과 비교
    for compare_col in all_cols.drop(missing_col):
        # 범주형 컬럼인 경우 그룹별 비율 확인
        if train_df[compare_col].dtype == 'object':
            print(f"'{compare_col}'에 따른 '{missing_col}' 결측치 비율:")
            # NaN 값은 groupby에 포함되지 않으므로, dropna=False로 전체를 대상으로 비율 계산
            grouped_data = train_df.groupby(compare_col)[f'{missing_col}_missing'].mean()
            print(grouped_data)
            print("-" * 30)
            
        # 연속형 컬럼인 경우 박스플롯으로 시각화
        elif train_df[compare_col].dtype in ['int64', 'float64']:
            plt.figure(figsize=(8, 6))
            sns.boxplot(x=f'{missing_col}_missing', y=compare_col, data=train_df)
            plt.title(f'{compare_col} Distribution by {missing_col} Missingness')
            plt.xticks([0, 1], ['Exists', 'Missing'])
            plt.show()

# 임시로 만든 결측치 컬럼 삭제
train_df.drop([f'{col}_missing' for col in missing_cols], axis=1, inplace=True)
```

```python
# 결측치 처리 함수 정의
def handle_missing_values(df):
    
    # 1. 'medical_history', 'family_medical_history' 결측치 처리
    # 'unknown'이라는 새로운 카테고리로 채우기
    df['medical_history'] = df['medical_history'].fillna('unknown')
    df['family_medical_history'] = df['family_medical_history'].fillna('unknown')
    
    # 2. 'edu_level' 결측치 처리
    # 'unknown'으로 채우기
    df['edu_level'] = df['edu_level'].fillna('unknown')
    
    # 3. 'mean_working' 결측치 처리
    # 'smoke_status'와 'edu_level'을 기준으로 그룹별 중앙값으로 채우기
    # 먼저 그룹별 중앙값 계산
    grouped_median = df.groupby(['smoke_status', 'edu_level'])['mean_working'].transform('median')
    # 계산된 중앙값으로 결측치 채우기
    df['mean_working'] = df['mean_working'].fillna(grouped_median)
    
    # 만약 그룹별 중앙값으로도 채워지지 않는 결측치가 있다면 (해당 그룹이 없는 경우), 전체 중앙값으로 채우기
    if df['mean_working'].isnull().sum() > 0:
        total_median = df['mean_working'].median()
        df['mean_working'] = df['mean_working'].fillna(total_median)
        
    return df

# 함수 적용
train_df = handle_missing_values(train_df)
test_df = handle_missing_values(test_df)

# 처리 후 결측치 확인
print("Train 데이터 결측치 처리 후:")
print(train_df.isnull().sum())
print("\nTest 데이터 결측치 처리 후:")
print(test_df.isnull().sum())
```

```python
# 'bone_density' 이상치 처리 함수
def handle_outliers(df):
    
    # 'bone_density' 컬럼에서 음수인 값들을 0으로 대체
    df.loc[df['bone_density'] < 0, 'bone_density'] = 0
    
    return df

# 함수 적용
train_df = handle_outliers(train_df)
test_df = handle_outliers(test_df)

# 처리 후 'bone_density'의 describe 결과 확인
print("Train 데이터 'bone_density' 처리 후:")
print(train_df['bone_density'].describe())
print("\nTest 데이터 'bone_density' 처리 후:")
print(test_df['bone_density'].describe())
```

```python
# 원-핫 인코딩 함수
def one_hot_encode(df):
    
    # 인코딩할 범주형 컬럼 리스트
    categorical_cols = ['gender', 'activity', 'smoke_status', 'sleep_pattern',
                        'medical_history', 'family_medical_history', 'edu_level']
    
    # 'unknown' 카테고리 포함 여부 확인
    print("원-핫 인코딩 전 'medical_history'의 유니크 값:")
    print(df['medical_history'].unique())

    # pd.get_dummies()를 사용하여 원-핫 인코딩
    df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
    
    return df

# 함수 적용
train_df = one_hot_encode(train_df)
test_df = one_hot_encode(test_df)

# 인코딩 후 데이터프레임의 컬럼 확인
print("\nTrain 데이터 인코딩 후 컬럼:")
print(train_df.columns)
print("\nTest 데이터 인코딩 후 컬럼:")
print(test_df.columns)
```

```python
# 'ID'와 'stress_score' 컬럼은 인덱스가 아니므로, False로 설정
train_df.to_csv('train_processed.csv', index=False)
test_df.to_csv('test_processed.csv', index=False)
```
