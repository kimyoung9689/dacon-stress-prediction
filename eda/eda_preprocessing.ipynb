```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
# 데이터 불러오기
train_df = pd.read_csv('../data/train.csv')
test_df = pd.read_csv('../data/test.csv')

# 데이터 기본 정보 확인
print("Train Data Info:")
train_df.info()

print("\nTest Data Info:")
test_df.info()

# 데이터 상위 5개 행 확인
print("\nTrain Data Head:")
print(train_df.head())
```

```python
# 기술 통계량 확인
print("Train Data Describe:")
print(train_df.describe())

print("\nTest Data Describe:")
print(test_df.describe())
```

```python
# 결측치 개수 확인
print("Train Data Missing Values:")
print(train_df.isnull().sum())

print("\nTest Data Missing Values:")
print(test_df.isnull().sum())
```

```python
# stress_score의 분포 확인 (히스토그램)
plt.figure(figsize=(10, 6))
sns.histplot(train_df['stress_score'], kde=True)
plt.title('Distribution of Stress Score')

# 이미지 저장
plt.savefig('images/stress_score_distribution.png')
plt.show() # 이미지가 저장된 후 화면에 보여주기

# 상관관계 히트맵
plt.figure(figsize=(12, 10))
sns.heatmap(train_df.corr(numeric_only=True), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap of Train Data')

# 이미지 저장
plt.savefig('images/correlation_heatmap.png')
plt.show()
```

```python
# 결측치 매트릭스 시각화
msno.matrix(train_df, figsize=(30, 15))
plt.title('Missing Value Matrix of Train Data', fontsize=15)
# 이미지 저장 코드를 먼저 실행
plt.savefig('images/missing_value_matrix.png')
plt.show() # 이미지가 저장된 후 화면에 보여주기

# 컬럼별 결측치 막대 그래프 시각화
msno.bar(train_df, figsize=(30, 15))
plt.title('Missing Value Bar Chart of Train Data', fontsize=15)
# 이미지 저장 코드를 먼저 실행
plt.savefig('images/missing_value_bar_chart.png')
plt.show()

# 결측치 상관관계 히트맵
msno.heatmap(train_df, figsize=(30, 15))
plt.title('Missing Value Correlation Heatmap', fontsize=15)
# 이미지 저장 코드를 먼저 실행
plt.savefig('images/missing_value_correlation_heatmap.png')
plt.show()
```

```python
# 결측치가 있는 4개 컬럼 리스트
missing_cols = ['medical_history', 'family_medical_history', 'edu_level', 'mean_working']
# 모든 컬럼 리스트 (ID와 stress_score는 제외하고)
all_cols = train_df.columns.drop(['ID', 'stress_score'])

# 각 결측치 컬럼에 대해 모든 컬럼과 비교 분석
for missing_col in missing_cols:
    print(f"\n--- {missing_col} 결측치 분석 ---")
    
    # 결측치 여부를 나타내는 컬럼 생성
    train_df[f'{missing_col}_missing'] = train_df[missing_col].isnull().astype(int)
    
    # 결측치 컬럼을 제외한 나머지 컬럼들과 비교
    for compare_col in all_cols.drop(missing_col):
        # 범주형 컬럼인 경우 그룹별 비율 확인
        if train_df[compare_col].dtype == 'object':
            print(f"'{compare_col}'에 따른 '{missing_col}' 결측치 비율:")
            grouped_data = train_df.groupby(compare_col)[f'{missing_col}_missing'].mean()
            print(grouped_data)
            print("-" * 30)
            
        # 연속형 컬럼인 경우 박스플롯으로 시각화
        elif train_df[compare_col].dtype in ['int64', 'float64']:
            plt.figure(figsize=(8, 6))
            sns.boxplot(x=f'{missing_col}_missing', y=compare_col, data=train_df)
            plt.title(f'{compare_col} Distribution by {missing_col} Missingness')
            plt.xticks([0, 1], ['Exists', 'Missing'])

            # 이미지 저장 코드
            image_path = f'images/boxplot_{missing_col}_{compare_col}.png' # 'EDA' -> 'eda'로 수정
            plt.savefig(image_path)
            
            plt.show()

# 임시로 만든 결측치 컬럼 삭제
train_df.drop([f'{col}_missing' for col in missing_cols], axis=1, inplace=True)
```
